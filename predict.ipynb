```python
import tensorflow as tf
from tensorflow.keras.applications.convnext import preprocess_input
import cv2
import numpy as np
import os

# Load trained model
model = tf.keras.models.load_model("sinhala_convnext_full.keras")

print("Model loaded successfully")
```

```python
CLASS_NAMES = sorted(os.listdir("dataset_processed"))
print("Classes:", CLASS_NAMES)
```

```python
IMG_SIZE = 128

def predict_image(img_path):
    img = cv2.imread(img_path)

    if img is None:
        raise ValueError("Image not found")

    # Resize
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))

    # Convert to float32
    img = img.astype(np.float32)

    # Apply ConvNeXt preprocessing
    img = preprocess_input(img)

    # Add batch dimension
    img = np.expand_dims(img, axis=0)

    # Predict
    preds = model.predict(img, verbose=0)
    class_index = np.argmax(preds)
    confidence = np.max(preds)

    predicted_label = CLASS_NAMES[class_index]

    return predicted_label, confidence
```

```python
SEGMENTED_DIR = "segmentation_output"

for file in os.listdir(SEGMENTED_DIR):
    img_path = os.path.join(SEGMENTED_DIR, file)

    label, conf = predict_image(img_path)

    print(f"{file} → {label} ({conf*100:.2f}%)")
```
